```python
from ultralytics import YOLO
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# ØªØ­Ù…ÙŠÙ„ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
model = YOLO("/content/runs/detect/train/weights/best.pt")  # ØºÙŠÙ‘Ø±ÙŠ Ø§Ù„Ù…Ø³Ø§Ø± Ù„Ùˆ Ø§Ø­ØªØ¬ØªÙ

# ØªØ­Ù…ÙŠÙ„ Ø§Ù„ØµÙˆØ±Ø©
image_path = "/content/free_kick2.png"  # ØºÙŠÙ‘Ø±ÙŠÙ‡Ø§ Ù„Ø§Ø³Ù… ØµÙˆØ±ØªÙƒ
img = cv2.imread(image_path)

# Ø¹Ù…Ù„ Ø§Ù„ØªÙ†Ø¨Ø¤
results = model(image_path)[0]

# Ø§Ø³ØªØ®Ø±Ø§Ø¬ Ø§Ù„Ø¨ÙˆÙƒØ³Ø§Øª Ù„Ù„ÙƒÙ„Ø§Ø³ÙŠÙ† Ø§Ù„Ù…Ø·Ù„ÙˆØ¨ÙŠÙ†
ball_box = None
space_box = None
for box, cls in zip(results.boxes.xyxy, results.boxes.cls):
    label = model.names[int(cls)]
    if label == "ball":
        ball_box = box
    elif label == "space":
        space_box = box

# Ø¥Ø°Ø§ ÙˆÙØ¬Ø¯ÙˆØ§ Ù†Ø­Ø³Ø¨ ÙˆÙ†Ø±Ø³Ù…
if ball_box is not None and space_box is not None:
    # Ø­Ø³Ø§Ø¨ Ø§Ù„Ù…Ø±ÙƒØ²
    bx1, by1, bx2, by2 = ball_box
    sx1, sy1, sx2, sy2 = space_box

    ball_center = ((bx1 + bx2) / 2, (by1 + by2) / 2)
    space_center = ((sx1 + sx2) / 2, (sy1 + sy2) / 2)

    # Ø­Ø³Ø§Ø¨ Ø§Ù„Ù…Ø³Ø§ÙØ©
    distance = int(np.linalg.norm(np.array(ball_center) - np.array(space_center)))

    # Ø±Ø³Ù… Ø§Ù„Ø®Ø·
    cv2.line(img, (int(ball_center[0]), int(ball_center[1])), (int(space_center[0]), int(space_center[1])), (255, 0, 0), 2)
    cv2.putText(img, f"Distance: {distance}px", (int(ball_center[0]), int(ball_center[1]) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2)

# Ø¹Ø±Ø¶ Ø§Ù„ØµÙˆØ±Ø©
cv2_imshow(img)
```
